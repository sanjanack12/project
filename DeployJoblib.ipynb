```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df=df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
lb=LabelEncoder()
lb1=LabelEncoder()
df['Sex']=lb.fit_transform(df['Sex'])
df['Embarked']=lb1.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
st=StandardScaler()
X_new=st.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.2,random_state=2)

```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
rf.score(X_train,y_train)
```

```python
import joblib
joblib.dump(rf,'Titanic.pkl')
joblib.dump(lb,'lb.pkl')
joblib.dump(lb1,'lb1.pkl')
joblib.dump(st,'st.pkl')
```
